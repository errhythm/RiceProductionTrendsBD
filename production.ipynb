```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import the dataset RiceProductionYearwise
df = pd.read_csv('RiceProductionYearwise.csv')
df
```

```python
#plot the graph where x axis is year and y axis is Aus, Aman, Boro, Total
# make the plot size big so that it covers all the year data
plt.figure(figsize=(50,20))
sns.lineplot(x='Year', y='Aus', data=df)
sns.lineplot(x='Year', y='Aman', data=df)
sns.lineplot(x='Year', y='Boro', data=df)
sns.lineplot(x='Year', y='Total', data=df)
plt.title('Rice Production Yearwise', fontsize=42)
plt.xlabel('Year')
plt.ylabel('Rice Production')
plt.legend(['Aus', 'Aman', 'Boro', 'Total'])
plt.savefig('images/RiceProductionYearwise.pdf')
plt.show()
```

```python
# import the riceprice data and riceprice_testdata and merge them in a df
riceprice_df = pd.read_csv('riceprice.csv')
riceprice_testdata_df = pd.read_csv('riceprice_test.csv')

# add riceprice_testdata_df to riceprice_df
price_df = riceprice_df.append(riceprice_testdata_df, ignore_index=True)
price_df
```

```python
# get mean data for every year for every rice types
price_df = price_df.groupby(['year', 'merchandise']).mean().reset_index()
price_df
```

```python
total_price_df = price_df.groupby('year').mean().reset_index()
total_price_df
```

```python
# make a lineplot where x axis is year and y axis is price when rice type is aman
plt.figure(figsize=(20,10))
sns.lineplot(x='year', y='price', data=price_df[price_df['merchandise']=='Aman'])
sns.lineplot(x='year', y='price', data=price_df[price_df['merchandise']=='Aus'])
sns.lineplot(x='year', y='price', data=price_df[price_df['merchandise']=='Boro'])
sns.lineplot(x='year', y='price', data=total_price_df)
# show all year
plt.xticks(price_df['year'].unique())
plt.title('Rice Price Yearwise', fontsize=42)
plt.xlabel('Year')
plt.ylabel('Rice Price')
plt.legend(['Aman', 'Aus', 'Boro', 'Average'])
plt.savefig('images/RicePriceYearwise.pdf')
plt.show()
```

```python
df = pd.read_csv('RiceProductionYearwise.csv')
df
```

```python
# fix the year column so that if it is 2019-20 it will be 2020
df['Year'] = df['Year'].apply(lambda x: int(x.split('-')[0])+1 if len(x.split('-'))>1 else int(x))
df
```

```python
# keep the data of year 2010 to 2020
df = df[(df['Year']>=2010) & (df['Year']<=2020)]
df
```

```python
# keep the data of price_df of year 2010 to 2020
new_price_df = price_df[(price_df['year']>=2010) & (price_df['year']<=2020)]
new_price_df
```

```python
# create a figure and axis
fig, ax1 = plt.subplots(figsize=(20,10))

# plot the production data on the first y-axis
sns.lineplot(x='Year', y='Aus', data=df, ax=ax1, color='blue')
ax1.set_xlabel('Year')
ax1.set_ylabel('Production (mt)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# create a second y-axis that shares the same x-axis
ax2 = ax1.twinx()

# plot the price data on the second y-axis
sns.lineplot(x='year', y='price', data=new_price_df[new_price_df['merchandise']=='Aus'], ax=ax2, color='red')
ax2.set_ylabel('Price (tk)', color='red')
ax2.tick_params(axis='y', labelcolor='red')
plt.xticks(new_price_df['year'].unique())
# add title and legend
plt.title('Rice Production and Price Yearwise (Aus)', fontsize=42)
plt.legend(['Production', 'Price'])

# save the plot as a PDF file
plt.savefig('images/RiceProductionAndPriceYearwise_Aus.pdf')

# show the plot
plt.show()
```
